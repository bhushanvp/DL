```python
import tensorflow as tf
from tensorflow.keras import layers, models
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/media/bhushan/Data/EDU/DL/Practicals/LP-IV-datasets/ECGdataset/ecg_autoencoder_dataset.csv', header=None)
df
```

```python
tr_l = df[140]
tr_d = df.drop([140], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(tr_d, tr_l)
x_train.shape
```

```python
input_layer = layers.Input(shape=(x_train.shape[1],))
encoded = layers.Dense(64, activation='relu')(input_layer)
encoded = layers.Dense(32, activation='relu')(encoded)
encoded = layers.Dense(16, activation='relu')(encoded)

# d. Decoder networks convert it back to the original input
decoded = layers.Dense(32, activation='relu')(encoded)
decoded = layers.Dense(64, activation='relu')(decoded)
decoded = layers.Dense(x_train.shape[1], activation='sigmoid')(decoded)
```

```python
autoencoder = models.Model(inputs=input_layer, outputs=decoded)
```

```python
autoencoder.compile(optimizer='adam', loss='mae')
```

```python
history = autoencoder.fit(x_train, x_train, epochs=10, batch_size=32, shuffle=True, validation_data=(x_test, x_test))
```

```python
reconstructed_data = autoencoder.predict(x_test)
reconstructed_data
```

```python
mae = np.mean(abs(x_test - reconstructed_data), axis=1)
mae
```

```python
y_pred = (mae>0.4)
```

```python
y_pred
```

```python
test_acc = np.mean(y_pred==y_test)
```
